```python
import cv2 
import os 
video = cv2.VideoCapture(r'/content/me-cuts/cuts/sub15-3.mp4') 
try:  
	if not os.path.exists('/content/gdrive/MyDrive/MEVIEW/data/neutral'): 
		os.makedirs('/content/gdrive/MyDrive/MEVIEW/data/neutral') 
except OSError: 
	print ('Error') 
currentframe = 0
while(True): 
	ret,frame = video.read() 

	if ret: 
		name = '/content/gdrive/MyDrive/MEVIEW/data/neutral/sub15-3-' + str(currentframe) + '.jpg'
		print ('Captured...' + name) 
		cv2.imwrite(name, frame) 
		currentframe += 1
	else: 
		break
video.release() 
cv2.destroyAllWindows()
```
