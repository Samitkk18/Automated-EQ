## kNN model for Audio dataset

```python
import os
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
```

Load Data

```python
df = pd.read_csv('Featuresets1/dataset.csv')
df
```

Drop file, start, end columns (unecessary)

```python
df = df.drop(['file', 'start', 'end'], axis ='columns')
df
```

Remove emotion column as output for the dataset

```python
X = df.drop(columns=['emotion'])
X.head()
```

Store the emotion column as a target column

```python
y = df['emotion']
y.to_frame()
```

```python
y = pd.get_dummies(y, columns=['emotion'])
y
```

Split training and testing dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)
```

Normalizing X_learn and X_test

```python
norm = MinMaxScaler(feature_range=(-1, 1)).fit(X_train)
X_train_norm = norm.transform(X_train)
X_test_norm = norm.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_norm, y_train)
```
