```python
data = pd.read_csv('dataset.csv')
```

```python
def get_file_emotion(file_path):
    # file_path = "data\\AudioData\\KL\\n17.wav"
    file_parts = file_path.split("\\")
    file_name = file_parts[-1]
    file_name = file_name.split('.')[0]
    # print(file_name)
    if 'a' in file_name:
        if 'sa' in file_name:
            return 'sadness'
        return 'anger'
    elif 'd' in file_name:
        return 'disgust'
    elif 'f' in file_name:
        return 'fear'
    elif 'h' in file_name:
        return 'happiness'
    elif 'n' in file_name:
        return 'neutral'
    else:
        return 'surprise'
```

```python
data['emotion'] = data['file'].apply(get_file_emotion)
data
```

```python
data = data.drop(['file', 'start', 'end'], axis='columns')
data
```

```python
Y = data['emotion'].to_frame()
Y
```

```python
X = data.drop(columns=['emotion'])
```

```python
norm = MinMaxScaler(feature_range=(-1, 1))
X_norm = norm.fit_transform(X)
X_norm
```

Feature Selection

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
threshold = 5
high_score_feats = []
y = np.ravel(Y.values)
feature_scores = mutual_info_classif(X_norm, y)
```

```python
feature_scores
```

```python
for score, f_name in sorted(zip(feature_scores, X.columns), reverse=True):
        print(f_name, score)
        if score > 0.0:
                high_score_feats.append(f_name)
```

```python
high_score_feats
```

```python
X_norm_df = pd.DataFrame(X_norm, columns=X.columns)
```

```python
data_mic = X_norm_df[high_score_feats]
```

```python
print(data_mic)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_mic, Y, test_size=0.2, random_state=27)
```
