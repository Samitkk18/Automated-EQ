Using openSmile to extract features

```python
import os
import time
import numpy as np
import pandas as pd
import audb
import audiofile
import opensmile
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
db = audb.load(name='emodb', version='1.1.1', format='wav', mixdown=True, sampling_rate=16000, media='wav/0.a01.*', full_path=False, verbose=False)
```

```python
smile = opensmile.Smile(feature_set=opensmile.FeatureSet.emobase, feature_level=opensmile.FeatureLevel.Functionals)
smile.feature_names
```

```python
files = db.files
df = smile.process_files(files, root = db.root)
```
